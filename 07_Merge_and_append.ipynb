```python
pd.concat([cost,sell])
```

```python
pd.merge(cost,sell , on = ['name' , 'name'],how="outer")
```

```python
pd.merge(cost,sell ,on=['name','name'])
```

```python
pd.merge(cost,sell ,on=['name','name'] , how = "inner")
```

```python
#full join(outer), lest join , right join,inner join
```

```python
pd.merge(cost,sell ,on=['name','name'] , how = "left")
```

## Side by side merge

```python
#full join(outer), lest join , right join,inner join
```

```python
pd.merge(cost,sell , on = ['name' , 'name'],how="outer")
```

```python
pd.merge(cost,sell ,on=['name','name'] , how = "inner")
```

```python
pd.merge(cost,sell ,on=['name','name'] , how = "left")
```

```python
pd.merge(cost,sell ,on=['name','name'] , how = "right")
```

## when columns are named differently

```python
pd.merge(cost,sell,left_on=['CP'],right_on=['SP'])
```
